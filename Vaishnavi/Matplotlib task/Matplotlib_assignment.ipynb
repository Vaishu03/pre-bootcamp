```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Scatterplot

```python
iris = pd.read_csv('/content/Iris.csv')
```

```python
iris.head()
```

```python
x = iris["SepalLengthCm"]
y = iris["PetalWidthCm"]
colors = iris["PetalLengthCm"]
size = iris["SepalWidthCm"]
```

```python
plt.title("Iris Dataset Scatterplot")
plt.xlabel("SepalLength")
plt.ylabel("PetalWidth")
plt.scatter(x,y,c = colors,s=size)
```

## Barplot

```python
Calories = np.random.random(10)
time_in_min = np.random.random(10)
x_axis = np.arange(len(time_in_min))
```

```python
plt.title("Calories burnt after workout session")
plt.xlabel("Time")
plt.ylabel("Calories")
plt.bar(x_axis,Calories,color="r")
```
